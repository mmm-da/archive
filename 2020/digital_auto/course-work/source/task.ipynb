# Задание по триггерам

```python
from automaton import MillyAutomaton,SUB

transition_matr = [ [2,1,4,4],
                    [4,3,2,3],
                    [4,4,3,1] ] 

output_matr = [ [1,2,2,1],
                [3,4,4,3],
                [2,4,3,4] ]

task = MillyAutomaton(transition_matr,output_matr)


print()
for index,i in enumerate(task.get_output_func(format='pretty'),start=1):
    print('Функция выходов Милли w{}:'.format(index).translate(SUB),i)
print()
for index,i in enumerate(task.get_D_trigger_func(format='pretty'),start=1):
    print('Функция возбуждения D-триггера f{}:'.format(index).translate(SUB),i)
print()
for index,i in enumerate(task.get_T_trigger_func(format='pretty'),start=1):
    print('Функция возбуждения T-триггера f{}:'.format(index).translate(SUB),i)
print()
for index,res in enumerate(task.get_RS_trigger_func(format='pretty'),start=1):
    print('Функция установки RS-триггера s{}:'.format(index).translate(SUB),res[0])
    print('Функция сброса RS-триггера r{}:'.format(index).translate(SUB),res[1])
print()
task.plot_graph()
```
