```python
from automaton import MillyAutomaton,Code,SUB

milly_transition_matr = [ [2,'-','-'],
                          [2,'-','-'],
                          [2,'-','-'],
                          ['-',2,'-'],
                          ['-',3,'-'],
                          ['-','-',3],
                          ['-','-',1] ] 

milly_output_matr = [     [2,'-','-'],
                          [3,'-','-'],
                          [2,'-','-'],
                          ['-',3,'-'],
                          ['-',5,'-'],
                          ['-','-',5],
                          ['-','-',1] ] 

output_codes = {}
output_codes[1]=Code(1,'-1-')
output_codes[2]=Code(2,'--1')
output_codes[3]=Code(3,'-11')
output_codes[4]=Code(4,'--1')
output_codes[5]=Code(5,'11-')

input_codes = {}
input_codes[1]=Code(1,'1--')
input_codes[2]=Code(2,'0-1')
input_codes[3]=Code(3,'0-0')
input_codes[4]=Code(4,'--1')
input_codes[5]=Code(5,'--0')
input_codes[6]=Code(6,'-0-')
input_codes[7]=Code(7,'-1-')



task = MillyAutomaton(milly_transition_matr,milly_output_matr)
task.output_codes = output_codes
task.input_codes = input_codes

print()
for index,i in enumerate(task.get_output_func(format = 'latex'),start=1):
    print('Функция выходов Милли $w_{}:$'.format(index),i)
print()
for index,i in enumerate(task.get_D_trigger_func(format = 'latex'),start=1):
    print('Функция возбуждения D-триггера $f_{}:$'.format(index),i)
print()
for index,i in enumerate(task.get_T_trigger_func(format = 'latex'),start=1):
    print('Функция возбуждения T-триггера $f_{}:$'.format(index),i)
print()

print()
for index,i in enumerate(task.get_output_func(format = 'lf'),start=1):
    print('Функция выходов Милли $w_{}:$'.format(index),i)
print()
for index,i in enumerate(task.get_D_trigger_func(format = 'lf'),start=1):
    print('Функция возбуждения D-триггера $f_{}:$'.format(index),i)
print()
for index,i in enumerate(task.get_T_trigger_func(format = 'lf'),start=1):
    print('Функция возбуждения T-триггера $f_{}:$'.format(index),i)
print()

print()
for index,i in enumerate(task.get_output_func(format = 'impl'),start=1):
    print('Функция выходов Милли $w_{}:$'.format(index),i)
print()
for index,i in enumerate(task.get_D_trigger_func(format = 'impl'),start=1):
    print('Функция возбуждения D-триггера $f_{}:$'.format(index),i)
print()
for index,i in enumerate(task.get_T_trigger_func(format = 'impl'),start=1):
    print('Функция возбуждения T-триггера $f_{}:$'.format(index),i)
print()
plot = task.plot_graph()
plot
```
