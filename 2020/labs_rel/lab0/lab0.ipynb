### Непрерывный случай

```python
import random
import numpy as np
lambda_val = 6/5
count_of_values = 500
arr = np.random.uniform(size=count_of_values)
print("n = ",count_of_values)
def xi2_kv(freq_t,freq_r):
    if(len(freq_r) != len(freq_t)):
        print("Количество теоретических и фактических частот отличается.")
        exit(1)
    kv = 0
    for i in range(0,len(freq_r)):
        kv = kv + ((freq_r[i]-freq_t[i])**2/freq_t[i])
    return kv
```

```python
import numpy as np
import matplotlib.pyplot as pp
pp.hist(arr)
pp.show()
```

Номер по списку $6$.
$\lambda = \frac{6}{5}$

```python
import math
import numpy
def to_cont_exp(val):
    return (-1/lambda_val)*numpy.log(val)
exp_cont_arr = []
for i in arr:
    exp_cont_arr.append(to_cont_exp(i))
```

```python
import numpy as np
import matplotlib.pyplot as pp
pp.hist(exp_cont_arr)
pp.show()
```

```python
import numpy as np
avg_exp_cont = np.mean(exp_cont_arr)
avg_exp_cont2= np.mean(np.square(exp_cont_arr))
print("Среднее арифметическое - ",avg_exp_cont)
print("Дисперсия - ",avg_exp_cont2-avg_exp_cont**2)
```

```python
import numpy as np 
step = (np.max(exp_cont_arr)-np.min(exp_cont_arr))/4
print("Шаг - ",step)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
int_count = [0,0,0,0]
int_Tcount = [0,0,0,0]
int_arr = []
for i in exp_cont_arr:
    if(0<i<=(step)):
        int_count[0]=int_count[0]+1
        int_arr.append(0)
    elif((step)<i<=(step*2)):
        int_count[1]=int_count[1]+1
        int_arr.append(1)
    elif((step*2)<i<=(step*3)):
        int_count[2]=int_count[2]+1
        int_arr.append(2)
    elif((step*3)<i<=step*4):
        int_count[3]=int_count[3]+1
        int_arr.append(3)
print(int_count)
int_Tcount[0] = (np.exp(-lambda_val*0) - np.exp(-lambda_val*step))*count_of_values
int_Tcount[1] = (np.exp(-lambda_val*step) - np.exp(-lambda_val*step*2))*count_of_values
int_Tcount[2] = (np.exp(-lambda_val*step*2) - np.exp(-lambda_val*step*3))*count_of_values
int_Tcount[3] = (np.exp(-lambda_val*step*3) - np.exp(-lambda_val*step*4))*count_of_values
print(int_Tcount)
plt.hist(int_arr,4)
plt.show()
print("Кв = ",xi2_kv(int_Tcount,int_count))
print("Табличное значение (3 степени свободы,уровень значимости 0.05) = 7.81")
```

### Дискретный случай

$$i = 0 , j = 4, k = 5, t = 2(i+j+k) = 18$$

|$x_i$| 0 | 1 | 2 | 3 |
|-----|---|---|---|---|
|$p_i$|$\frac{i}{t}$|$\frac{j}{t}$|$\frac{i+j}{t}$|$\frac{2k}{t}$|


|$x_i$| 0 | 1 | 2 | 3 |
|-----|---|---|---|---|
|$p_i$|$\frac{1}{18}$|$\frac{3}{18}$|$\frac{4}{18}$|$\frac{10}{18}$|

```python
import numpy as np 
import matplotlib.pyplot as plt
v_count = [0,0,0,0]
v_Tcount = [1/18*count_of_values,3/18*count_of_values,4/18*count_of_values,10/18*count_of_values]
v_arr = []
for i in arr:
    if(0<i<=(1/18)):
        v_count[0]=v_count[0]+1
        v_arr.append(0)
    elif((1/18)<i<=(4/18)):
        v_count[1]=v_count[1]+1
        v_arr.append(1)
    elif((4/18)<i<=(8/18)):
        v_count[2]=v_count[2]+1
        v_arr.append(2)
    elif((8/18)<i<=1):
        v_count[3]=v_count[3]+1
        v_arr.append(3)
print(v_count)
print(v_Tcount)
plt.hist(v_arr,4)
plt.show()
discrete_mean = np.mean(v_arr)
discrete_mean2 = np.mean(np.square(v_arr))
dispertion = discrete_mean2 - discrete_mean**2 
print("Среднее арифметическое - ",discrete_mean)
print("Дисперсия  - ",dispertion)
print("Кв = ",xi2_kv(v_Tcount,v_count))
print("Табличное значение (3 степени свободы,уровень значимости 0.05) = 7.81")
```
